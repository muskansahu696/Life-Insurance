```python
sns.distplot(data["Ins_Age"])
plt.show()
```

```python
def bucke(x):
    if (x>=0) & (x<=0.2):
        return 0
    elif (x>.2) & (x<=.4):
        return 1
    elif (x>.4) & (x<=.6):
        return 2
    elif (x>.6) & (x<=.8):
        return 3
    else:
        return 4
data["buk_Ins_age"]=data["Ins_Age"].apply(bucke)
```
